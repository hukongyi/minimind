```python
import json
import os

# --- 配置 ---
file_path = "../dataset/pretrain_hq.jsonl"
# 设置你想查看的最多行数，现在固定为 100
max_lines_to_show = 100

# --- 主逻辑 ---
if not os.path.exists(file_path):
    print(f"错误：文件不存在于路径 '{file_path}'")
else:
    print(f"--- 开始读取文件 (最多显示前 {max_lines_to_show} 行): {file_path} ---")
    try:
        line_count = 0
        with open(file_path, "r", encoding="utf-8") as f:
            for line in f:
                # 检查是否已经达到需要显示的行数限制（在处理当前行之前检查）
                # 这样可以确保我们准确地处理第 1 到 第 100 行
                if line_count >= max_lines_to_show:
                    print(f"\n--- 已显示前 {max_lines_to_show} 行，停止读取 ---")
                    break  # 跳出 for 循环

                line_count += 1  # 行数计数器加 1 (代表这是第 line_count 行)
                line = line.strip()  # 去除行首尾的空白符（包括换行符）

                print(f"--- 行 {line_count} ---")

                if not line:  # 跳过空行，但仍然计数
                    print("(空行)")
                    continue  # 继续下一轮循环

                try:
                    # 解析当前行的 JSON 字符串为 Python 对象 (通常是字典)
                    data = json.loads(line)

                    # 打印解析后的 Python 对象 (为了更好的可读性，使用 json.dumps 格式化输出)
                    print(json.dumps(data, indent=2, ensure_ascii=False))

                except json.JSONDecodeError as e:
                    print(f"错误：无法解析此行为 JSON: {e}")
                    print(f"原始行内容: {line}")

        # 如果文件总行数少于 max_lines_to_show，循环会自然结束
        if line_count < max_lines_to_show:
            print(f"\n--- 文件读取结束 (文件总共只有 {line_count} 行) ---")
        # else: # 如果是因为达到 100 行而 break 的，上面的 break 处已经打印了提示信息
        #     pass

    except FileNotFoundError:
        # 这个检查理论上在上面 os.path.exists 已经处理了，但为了健壮性再加一层
        print(f"错误：再次确认，文件未找到 '{file_path}'")
    except Exception as e:
        print(f"读取或处理文件时发生意外错误: {e}")
```
